```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hotels.csv')
```

```python
df.drop(['company'], axis=1, inplace=True)
```

```python
df['agent'].fillna(0, inplace=True)
```

```python
df['children'].fillna(0, inplace=True)
```

```python
df['country'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
filter_data = (df['adults'] + df['children'] + df['babies'] > 0)

df = df[filter_data]
```

```python
df['children'] = df['children'].astype(int)
df['agent'] = df['agent'].astype(int)
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

df['reservation_year'] = df['reservation_status_date'].dt.year
df['reservation_month'] = df['reservation_status_date'].dt.month
df['reservation_day'] = df['reservation_status_date'].dt.day
```

```python
df.isnull().sum()
```

```python
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
df['has_kids'] = ((df['children'] + df['babies']) > 0).astype(int)
```

```python
df[['adults', 'children', 'babies', 'total_guests', 'total_stay', 'has_kids']].head()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='hotel', hue='is_canceled', data=df)
plt.title('Otel Tipinə Görə Ləğv Edilmə Halları')
plt.xlabel('Otel Tipi')
plt.ylabel('Rezervasiya Sayı')
plt.legend(title='Ləğv', labels=['Xeyr', 'Bəli'])

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='market_segment', data=df, order=df['market_segment'].value_counts().index)
plt.title('Rezervasiya Mənbələrinin Paylanması')
plt.xlabel('Mənbə')
plt.ylabel('Sayı')
plt.xticks(rotation=45)

plt.show()
```

```python
top_10_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values)
plt.title('Ən Çox Qonaq Gələn İlk 10 Ölkə')
plt.xlabel('Ölkə Kodu')
plt.ylabel('Qonaq Sayı')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='lead_time', hue='is_canceled', fill=True)
plt.title('Gözləmə Müddətinin Ləğv Edilməyə Təsiri')

plt.show()
```

```python
deposit_cancel = pd.crosstab(df['deposit_type'], df['is_canceled'], normalize='index')
deposit_cancel.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Depozit Növünə Görə Ləğv Faizi')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.pointplot(x='has_kids', y='total_of_special_requests', hue='hotel', data=df)
plt.title('Uşaqlı və Uşaqsız Qonaqların Xüsusi İstək Ortalaması')

plt.show()
```
