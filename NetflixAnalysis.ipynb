```python
# In three rows of the dataset, the movie's duration (such as 90 min) was mistakenly entered into the age rating column, leaving the actual duration column blank.
# This would disrupt both duration calculations and rating groupings in future analyses.
```

```python
df['rating'] = df.loc[df['duration'].isnull(), 'duration']

df.loc[df['rating'].str.contains('min', na=False), 'rating'] = 'Unknown'
```

```python
# We found the hidden duration information in the rating column and moved it to its correct duration column.
# After recovering the information, we cleaned up the incorrect "minutes" remaining in the rating column and replaced them with "Unknown".
```

```python
df.isna().sum()
```

```python
df['rating'].fillna('Unknown', inplace=True)
```

```python
df.loc[df['show_id'] == 's5542', 'duration'] = '74 min'
df.loc[df['show_id'] == 's5795', 'duration'] = '84 min'
df.loc[df['show_id'] == 's5814', 'duration'] = '66 min'
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='type', data=df, palette='viridis')
plt.title('Movie vs TV Show Paylanması')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df['year_added'].value_counts().sort_index())
plt.title('İllər üzrə Əlavə Olunan Kontent Sayı')
plt.ylabel('Sayı')

plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title('Ən Çox Kontent İstehsal Edən 10 Ölkə')

plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()
movies['duration_min'] = movies['duration'].str.replace('min', '').astype(float)

plt.figure(figsize=(10, 6))
sns.histplot(movies['duration_min'], bins=30, kde=True)
plt.title('Filmlərin Müddət (Dəqiqə) Paylanması')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=35, color='orange')
plt.title('Kontentlərin Orijinal Buraxılış İlləri')

plt.show()
```

```python
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(12, 6))
sns.countplot(x='month_added', data=df, order=months_order, palette='Spectral')
plt.title('Aylar üzrə Əlavə Edilən Kontent')

plt.show()
```

```python
top_5_countries = df[df['country'].isin(df['country'].value_counts().head(5).index)]

plt.figure(figsize=(10, 6))
sns.countplot(x='country', hue='type', data=top_5_countries)
plt.title('Top 5 Ölkədə Movie vs TV Show')

plt.show()
```

```python
directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=directors.values, y=directors.index, palette='coolwarm')
plt.title('Ən Çox Filmi/Serialı Olan 10 Rejissor')

plt.show()
```
